```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
print(load_dotenv())
```

```python
model = ChatOpenAI()
```

```python
class QAllm(TypedDict):
    question: str
    answer: str
```

```python
# creating graph object:
graph = StateGraph(QAllm)
```

```python
def llmqa(state: QAllm) -> QAllm:
    question = state["question"]
    prompt = f"Answer the followng question: {question}"
    answer = model.invoke(prompt).content
    state["answer"] = answer
    return state
```

```python
# add nodes:
graph.add_node("prompt_processing", llmqa)
# add edges:
graph.add_edge(START, "prompt_processing")
graph.add_edge("prompt_processing", END)
```

```python

# compile graph:
workflow = graph.compile()
```

```python

# run graph:
question = input("Ask Question: ")
print(question)
```

```python

questin_dict = {"question": question}
output = workflow.invoke(questin_dict)
print(output)
```
