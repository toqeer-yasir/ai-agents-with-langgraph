```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
pdf = "test_document.pdf"
```

```python
def load_and_chunk(doc_path):
    loader = PyPDFLoader(doc_path)
    doc_text = loader.load()

    splitter = RecursiveCharacterTextSplitter(
        chunk_size= 500,
        chunk_overlap= 100,
        separators= ["\n\n", "\n", " ", ""] 
    )
    chunks_doc = splitter.split_documents(doc_text)
    return chunks_doc
```

```python
print(len(load_and_chunk(pdf)))
print(load_and_chunk(pdf)[3])
```

```python
def embedd_and_store(chunks):
    embeddings = HuggingFaceEmbeddings(
        model_name="BAAI/bge-small-en-v1.5"
        )
    vector_store = Chroma.from_documents(
        documents= chunks,
        embedding= embeddings,
        persist_directory= "./chroma_db"
        )
    return vector_store
```

```python
vector_store = embedd_and_store(load_and_chunk(pdf))
```

```python
query = "What is the best way to write a resume?"
```

```python
retriever = vector_store.as_retriever(search_kwargs= {"k": 3})
docs = retriever.invoke(query)
```
